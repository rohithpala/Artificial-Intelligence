```python
import pandas as pd
```

```python
data = pd.read_csv("iris.csv")
test_set = [7.2, 3.6, 5.1, 2.5]
k = 10
data
```

```python
noOfInstances, noOfAttributes = data.shape
noOfAttributes -= 1
print("No. of Instances:", noOfInstances, "\nNo. of Attributes:", noOfAttributes)
```

```python
instances = data.iloc[:, :-1]
target_values = data.iloc[:, -1]
print("Instances:\n", instances)
print("Target Values:\n", target_values)
```

```python
def euclideanDistance(d1, d2, l):
    distance = 0
    for i in range(l):
        distance += (d1[i]-d2[i])**2
    return distance
```

### KNN Code

```python
distances = []
length = len(test_set)
for i in range(noOfInstances):
    distances.append(euclideanDistance(instances.loc[i], test_set, length))

# distances
distKeys = list(enumerate(distances))
distKeys.sort(key=lambda x: x[1])

distKeys
```

```python
kNearestNeighbours = distKeys[:k]
kNearestNeighbours
```

```python
classes = list(set(target_values))
classes
```

```python
# voting
votes = dict()

for class_ in classes:
    votes[class_] = 0

for i in range(k):
    label = str(target_values[kNearestNeighbours[i][0]])
    votes[label] = votes.get(label, 0) + 1

print(votes)
print("The given test set belongs to \"" + max(votes) + "\"")
```
